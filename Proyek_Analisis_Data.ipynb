### Cleaning Data

```python
#Pada kolom PM2.5, PM10, SO2, NO2, CO, dan O3 menggunakan teknik interpolate dikarenakan polutan ini cenderung berkaitan satu sama lain
dongsi_data_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']] = dongsi_data_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].interpolate(method='linear')
```

```python
#Pada kolom TEMP, PRES, DEWP, dan RAIN menggunakan teknik interpolate dikarenakan kondisi cuaca ini cenderung berkaitan satu sama lain
dongsi_data_df[['TEMP', 'PRES', 'DEWP', 'RAIN']] = dongsi_data_df[['TEMP', 'PRES', 'DEWP', 'RAIN']].interpolate(method='linear')
```

```python
#Teknik imputation digunakan pada kolom wd dengan Mode(data yang sering muncul)
dongsi_data_df['wd'] = dongsi_data_df['wd'].fillna(dongsi_data_df['wd'].mode()[0])
```

```python
#Pada kolom WSPM menggunakan teknik interpolate karena kecepatan angin secara kontinu berubah
dongsi_data_df['WSPM'] = dongsi_data_df['WSPM'].interpolate(method='linear')
```

**Insight:**
- Pada kolom wd digunakan teknik Impute mode karena mewakili nilai yang paling umum
- Teknik interpolate digunakan karena data pada kolom tersebut saling berhubungan

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada pukul berapakah rata rata konsentrasi CO pada udara tinggi?

```python
#Menghitung rata-rata kadar CO per jam
hourly_co = dongsi_data_df.groupby(by=["hour"]).agg({
    "CO": "mean"
})

#Membuat plot
hourly_co.plot(kind="bar", color="g", figsize=(10,5))
plt.title("Rata-rata Konsentrasi CO per Jam")
plt.xlabel("Jam")
plt.ylabel("Rata-rata Konsentrasi CO (ppm)")
plt.xticks(rotation=0)
plt.show()
```

### Pertanyaan 2: Faktor apa saja yang mempengaruhi atau berhubungan dengan konsentrasi CO?

```python
# Menghitung hubungan dengan metode pearson
corr_dongsi = dongsi_data_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].corr(method='pearson')

#Membuat Heatmap
sns.heatmap(corr_dongsi, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Pollutants (Pearson)')
plt.show()
```

**Insight:**
- Pada Bar Chart pertama menunjukan bahwa rata-rata kadar CO tinggi mulai pukul 23-02 dengan kadar oksigen tertinggi terdapat pada pukul 2 pagi
- Pada Heatmap korelasi ditunjukan bahwa terdapat hubungan yang kuat antara kenaikan kadar CO di udara dengan PM2.5, PM10, SO2, dan NO2
